```python
data.isnull().sum()
```

```python
data = data.apply(lambda x: x.fillna(x.median()), axis=0)
```

```python
data.isnull().sum()
```

```python
x = data.drop('class',axis=1)
y = data['class']
```

#### Data Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scale = x.copy()
scaled = sc.fit_transform(scale)
scaled_df = pd.DataFrame(scaled,columns=x.columns)
```

```python
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(scaled_df,y,train_size=0.70,random_state=1000)
```

#### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc_model = rfc.fit(xtrain,ytrain)
```

```python
test_pred = rfc_model.predict(xtest)
train_pred = rfc_model.predict(xtrain)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(ytest,test_pred))
```

```python
print(classification_report(ytrain,train_pred))
```

```python
confusion_matrix(ytest,test_pred)
```

```python
confusion_matrix(ytrain,train_pred)
```
